```python
import pandas as pd 
#Reading Data

df1 = pd.read_csv("/Users/USUARIO/Desktop/Data Analysis/Ironhack/Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file1.csv")
df2 = pd.read_csv("/Users/USUARIO/Desktop/Data Analysis/Ironhack/Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file2.csv")
df3 = pd.read_csv("/Users/USUARIO/Desktop/Data Analysis/Ironhack/Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv")

display(df1)
display(df2)
display(df3)
```

```python
df_list = [df1, df2, df3]

# Showing dataframes shapes

for i in df_list:
    print(i.shape)
```

```python
# Standardizing Header Names
for i in df_list:
        i.columns = i.columns.str.upper().str.replace(" ","_")
        
display(df1)
display(df2)
display(df3)
```

```python
# Changing the name of a column
df3 = df3.rename(columns={'STATE': 'ST'})

#Rearranging columns 

df1 = df1[["CUSTOMER", "ST", "GENDER", "EDUCATION", "INCOME", "CUSTOMER_LIFETIME_VALUE","NUMBER_OF_OPEN_COMPLAINTS", "VEHICLE_CLASS", "POLICY_TYPE", "MONTHLY_PREMIUM_AUTO", "TOTAL_CLAIM_AMOUNT"]]
df2 = df2[["CUSTOMER", "ST", "GENDER", "EDUCATION", "INCOME", "CUSTOMER_LIFETIME_VALUE","NUMBER_OF_OPEN_COMPLAINTS", "VEHICLE_CLASS", "POLICY_TYPE", "MONTHLY_PREMIUM_AUTO", "TOTAL_CLAIM_AMOUNT"]]
df3 = df3[["CUSTOMER", "ST", "GENDER", "EDUCATION", "INCOME", "CUSTOMER_LIFETIME_VALUE","NUMBER_OF_OPEN_COMPLAINTS", "VEHICLE_CLASS", "POLICY_TYPE", "MONTHLY_PREMIUM_AUTO", "TOTAL_CLAIM_AMOUNT"]]

display(df1)
display(df2)
display(df3)
```

```python
#Concatenating dataframes

data = pd.concat([df1,df2,df3], axis=0).reset_index(drop=True)
data
```

```python
#Numerical variables

numerical_vars = data[["CUSTOMER_LIFETIME_VALUE", "INCOME", "MONTHLY_PREMIUM_AUTO", "NUMBER_OF_OPEN_COMPLAINTS","TOTAL_CLAIM_AMOUNT"]]
numerical_vars
```

```python
#Categorical variables

categorical_vars = data[["CUSTOMER", "ST", "GENDER", "EDUCATION", "POLICY_TYPE", "VEHICLE_CLASS"]]
categorical_vars
```

```python
data = data.drop(["EDUCATION", "NUMBER_OF_OPEN_COMPLAINTS"], axis=1)
data
```

```python
#Changing Customer type to numeric values 
data["CUSTOMER_LIFETIME_VALUE"] = pd.to_numeric(data["CUSTOMER_LIFETIME_VALUE"], errors="coerce")
data

#Multiply *100 

data["CUSTOMER_LIFETIME_VALUE"] = data["CUSTOMER_LIFETIME_VALUE"] * 100
data
```

```python
#Cleaning NaNs

cleaned_data = data.dropna()
cleaned_data
```

```python
#Cleaning duplicates

cleaned_data = cleaned_data.drop_duplicates()
cleaned_data
```
